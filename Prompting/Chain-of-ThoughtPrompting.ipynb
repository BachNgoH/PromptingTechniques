```python
from openai import OpenAI

client = OpenAI(
  api_key='YOUR_API_KEY',
)
```

```python
def get_gpt_response(prompt, max_tokens=200):
    response = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=[{"role": "user", "content": prompt}],
        # max_tokens=500,
        # n=1,
        # stop="\n",  # Set a stop sequence here
        # temperature=0.7,
    )
    return response.choices[0].message.content
```

# **Prompting**

```python
# Standard Prompting
prompt = '''
Câu hỏi: A có 5 cái kẹo. A được bố và mẹ cho thêm kẹo, mỗi người cho thêm 3 viên. Hỏi bây giờ A có tất cả bao nhiêu cái kẹo?
Trả lời: Đáp án là 11

Câu hỏi: Một căn tin đang có 23 quả táo. Nếu họ sử dụng 20 quả để làm bữa trưa và mua thêm 6 quả nữa thì họ có tất cả bao nhiêu quả táo?
'''
response = get_gpt_response(prompt)
print(response)
```

```python
# Chain-of-Thought Prompting
prompt = '''
Câu hỏi: A có 5 cái kẹo. A được bố và mẹ cho thêm kẹo, mỗi người cho thêm 3 viên. Hỏi bây giờ A có tất cả bao nhiêu cái kẹo?
Trả lời: A ban đầu có 5 cái kẹo. Bố mẹ mỗi người cho thêm 3 viên là tổng cộng 6 viên. 5 + 6 = 11. Đáp án là 11 viên.

Câu hỏi: Một căn tin đang có 23 quả táo. Nếu họ sử dụng 20 quả để làm bữa trưa và mua thêm 6 quả nữa thì họ có tất cả bao nhiêu quả táo?
'''
response = get_gpt_response(prompt)
print(response)
```
