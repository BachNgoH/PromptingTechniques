# Init conversation

```python
from openai import OpenAI
import os


def get_completion(prompt, params):
    messages = [{"role": "user", "content":prompt}]
    client = OpenAI(
      api_key=os.environ.get("OPENAI_API_KEY"),
    )
    response = client.chat.completions.create(
        **params,
        messages=messages
    )

    return response.choices[0].message.content


os.environ['OPENAI_API_KEY'] = "YOUR_API_KEY"

modelName = 'gpt-3.5-turbo'
temperature = 0.7
top_p = 1

params = {
    "model": modelName,
    "temperature": temperature,
    "top_p": top_p
}
```

```python
# Without Tree of Thought
prompt = '''
An cầm một quả bóng trên tay. Sau đó, An đưa quả bóng cho Linh. Cuối cùng, Linh đưa cho Thanh một quả bóng và Linh trả cho An một quả quả bóng. Hỏi có bao nhiêu quả bóng trong câu chuyện
'''
response = get_completion(prompt, params)
print(response)
```

```python
# Tree of Thought
prompt = '''
Hãy tưởng tượng ba chuyên gia khác nhau đang trả lời câu hỏi này.
Tất cả các chuyên gia sẽ viết ra 1 bước suy nghĩ của mình, sau đó chia sẻ nó với nhóm.
Sau đó tất cả các chuyên gia sẽ chuyển sang bước tiếp theo, v.v.
Nếu bất kỳ chuyên gia nào nhận ra họ sai ở bất kỳ điểm nào thì họ sẽ rời đi.
Q:
An cầm một quả bóng trên tay. Sau đó, An đưa quả bóng cho Linh. Cuối cùng, Linh đưa cho Thanh một quả bóng và Linh trả cho An một quả quả bóng. Hỏi có bao nhiêu quả bóng trong câu chuyện
'''
response = get_completion(prompt, params)
print(response)
```
