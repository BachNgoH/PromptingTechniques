```python
from openai import OpenAI

client = OpenAI(
  api_key='YOUR_API_KEY',
)
```

```python
def get_gpt_response(prompt, max_tokens=200):
    response = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=[{"role": "user", "content": prompt}],
        # max_tokens=500,
        # n=1,
        # stop="\n",  # Set a stop sequence here
        # temperature=0.7,
    )
    return response.choices[0].message.content
```

# **Prompting**

```python
prompt = '''
Câu hỏi: Một cửa hàng thú cưng có 64 chú cún con. Một ngày, họ bán đi 28 con và chia số còn lại vào các lồng với mỗi lồng chứa 4 con.
Hỏi cửa hàng đó cần bao nhiêu lồng.
Trả lời: Hãy suy nghĩ từng bước.
'''
response = get_gpt_response(prompt)
print(response)
```
