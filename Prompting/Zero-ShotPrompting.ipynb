```python
from openai import OpenAI

client = OpenAI(
  api_key='YOUR_API_KEY',
)
```

```python
def get_gpt_response(prompt, max_tokens=200):
    response = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=[{"role": "user", "content": prompt}],
        # max_tokens=500,
        # n=1,
        # stop="\n",  # Set a stop sequence here
        # temperature=0.7,
    )
    return response.choices[0].message.content
```

# **Prompting**

```python
prompt = '''
Bạn là một chuyên gia phân tích email phản hồi của khách hàng. Bạn hãy giúp tôi đánh giá thái độ nhân viên qua mail
phản hồi cho khách hàng đây là thái độ tích cực hay tiêu cực hay bình thường:
"Dear Mr.Anh,

Thực lòng xin lỗi ông vì đã để ông phải dành thời gian với đội Sale của chúng tôi hơn 20 phút. Điều này sẽ
không bao giờ xảy ra nữa, và tôi hoàn toàn hiểu được điều đó gây thất vọng cho ông như thế nào.

Chúng tôi đang xử lý những sự cố trong hệ thống và sẽ thông báo cho ông ngay sau khi sự cố được giải quyết.
Chúng tôi sẽ cố gắng đảm bảo cho mỗi khách hàng đều hài lòng với doanh nghiệp và chúng tôi xin lỗi về những bất tiện đã gây ra cho ông.
Hãy cho chúng tôi biết nếu ông còn những câu hỏi, ý kiến, băn khoăn hoặc sự quan tâm nào với chúng tôi.

Trân trọng,

Mark.
'''
response = get_gpt_response(prompt)
print(response)
```

**Response:** Thái độ của nhân viên qua email phản hồi này có thể được đánh giá là tích cực. Nhân viên đã tỏ ra lịch sự, chân thành và biết lắng nghe ý kiến của khách hàng. Việc xin lỗi và cam kết sửa đổi để tránh tái diễn sai sót là điều tích cực. Họ cũng đã mô tả rõ lý do của sự cố và cam kết sẽ thông báo cho khách hàng sau khi vấn đề được giải quyết. Điều này cho thấy họ chú trọng đến sự hài lòng của khách hàng và sẵn lòng hỗ trợ khách hàng nếu có bất kỳ thắc mắc nào. Tóm lại, thái độ của nhân viên qua email là tích cực và chuyên nghiệp.
