# Init conversation

```python
from openai import OpenAI
import os


def get_completion(prompt, params):
    messages = [{"role": "user", "content":prompt}]
    client = OpenAI(
      api_key=os.environ.get("OPENAI_API_KEY"),
    )
    response = client.chat.completions.create(
        **params,
        messages=messages
    )

    return response.choices[0].message.content


os.environ['OPENAI_API_KEY'] = "YOUR_API_KEY"

modelName = 'gpt-3.5-turbo'
temperature = 0.7
top_p = 1

params = {
    "model": modelName,
    "temperature": temperature,
    "top_p": top_p
}
```

```python
# Without Self-consistency
prompt = '''Khi tôi 6 tuổi thì chị tôi gấp 2 lần tuổi của tôi. Khi tôi 10 tuổi thì chị tôi bao nhiêu tuổi'''
response = get_completion(prompt, params)
print(response)
```

```python
# Self-consistency
prompt = '''
Q: Bạn Lan có 10 cái kẹo, Lan cho Linh 6 cái kẹo. Hỏi Lan còn bao nhiêu cái kẹo?
A: Bạn Lan ban đầu có 10 cái kẹo sau đó cho Linh 6 cái thì số kẹo Lan còn lại là 10 - 6 = 4 cái kẹo. Lan còn lại 4 cái kẹo.
Q: Trong vườn có 15 cây chuối, một ông lão vào vườn trồng cây và đến cuối ngày, số cây chuối trong vườn là 20. Ông lão trồng được mấy cây chuối
A: Ban đầu trong vườn có 15 cây, sau khi ông lão trồng cây thì số cây trong vườn là 20. Suy ra số cây ông lão trồng được là 20 - 15 là 5 cây.
Q: Linh có 20 ngàn đồng. Giá của một cây kẹo là 2 ngàn một cái. Linh mua 5 cái kẹo. Linh còn lại bao nhiêu tiền.
A: Linh mua 5 cái kẹo với giá 2 ngàn cho 1 cái kẹo. Suy ra số tiền Linh mua kẹo là 2 * 5 = 10 ngàn. Linh còn lại số tiền là 20 - 10 = 10 ngàn.
Q: Trong nông trại có 2 người làm, người đầu tiên làm được 100 mẫu đất, người thứ hai làm được gấp 3 lần người đầu tiên. Hỏi tổng số mẫu đất 2 người làm được là bao nhiêu
A: Người thứ nhất làm được 100 mẫu đất, người thứ hai làm gấp ba lần người thứ nhất có nghĩa là người thứ hai làm được 100 x 3 = 300 mẫu đất. Vậy tổng mẫu đất hai người làm được là 100 + 300 = 400 mẫu đất
Q: Khi tôi 6 tuổi thì chị tôi gấp 2 lần tuổi của tôi. Khi tôi 10 tuổi thì chị tôi bao nhiêu tuổi
A:
'''
response = get_completion(prompt, params)
print(response)
```

Output 3: Khi bạn 6 tuổi, chị bạn gấp 2 lần tuổi của bạn, nghĩa là chị bạn là 6 x 2 = 12 tuổi. Khi bạn 10 tuổi, chị bạn sẽ còn thêm 10 - 6 = 4 năm, vì vậy chị bạn sẽ là 12 + 4 = 16 tuổi khi bạn 10 tuổi.

```python
# Self-consistency
prompt = '''
Output 1: Khi bạn 6 tuổi, chị bạn gấp 2 lần tuổi của bạn, tức là chị bạn là 6 x 2 = 12 tuổi. Khi bạn 10 tuổi, chị bạn sẽ lớn hơn bạn 4 tuổi, vì vậy chị bạn sẽ là 10 + 4 = 14 tuổi.
Output 2: Khi bạn 6 tuổi, chị của bạn gấp 2 lần tuổi của bạn, nghĩa là chị của bạn là 6 x 2 = 12 tuổi. Khi bạn 10 tuổi, chị của bạn cách bạn 4 tuổi, vì vậy chị của bạn khi bạn 10 tuổi sẽ là 12 + 4 = 16 tuổi.
Output 3: Khi bạn 6 tuổi, chị bạn gấp 2 lần tuổi của bạn, nghĩa là chị bạn là 6 x 2 = 12 tuổi. Khi bạn 10 tuổi, chị bạn sẽ còn thêm 10 - 6 = 4 năm, vì vậy chị bạn sẽ là 12 + 4 = 16 tuổi khi bạn 10 tuổi.
Q: Cuối cùng chị tôi bao nhiêu tuổi
'''
response = get_completion(prompt, params)
print(response)
```
